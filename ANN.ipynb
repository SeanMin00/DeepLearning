### 2. Making simple Artificial Neural Network (ANN) using ML method


> Topic: Tried to build ANN about the fomula, Y = 4X + 1, through ML method. The goal is getting output Y is 41 when input X is 10, by trained ANN


*   Imported *Numpy* library to make array, using `np.array` function. Input for this fuction is List.
*   Imported *Tensorflow* library to make ANN, using Keras classes.
1.   `tf.keras.Sequential` : use **Sequential model**  for the most simple ANN (single input & output tensor)
2.   `tf.keras.layers.InputLayer` : this function creates tensor provided by Keras, and set how many inputs are needed
3.   `tf.keras.layers.Dense` : this class make ANN layer. (Y = X*w + b here)




*   Imported below functions
1.   `model.compile` : requires `loss` funtion (using `mean_squared_error`) and `optimizer` function (`sgd`: stochastic gradient descent)
2.   `model.fit` : Let ANN study itself. `ephochs` = how many times of studies are required
3. `model.predict`: input value for ANN function



```python
# Goal: Make Y = 4X + 1 ANN, input is 10 and output is 41.
import numpy as np
import tensorflow as tf

# for more detetails
from tensorflow import keras
from tensorflow.keras import layers

# set the input&ouput for the function, Y = 4X + 1
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
ys = np.array([-3.0, 1.0, 5.0, 9.0, 13.0, 17.0])

# build ANN model using keras
model = keras.Sequential(
    [
        keras.layers.InputLayer(input_shape=(1,)),
        keras.layers.Dense(1)
     ]
)

#apply functions
model.compile(optimizer='sgd',
              loss='mean_squared_error')

model.fit(xs, ys, epochs=500)

p = model.predict([10.0])

# Let's get the ouput of 41
print('p:', p)
```

### 3. Generate and predict Quaratic function using ANN



> Topic: Generated ANN model prediting quadratic function

**0. Setup**

*   Imported *Time* to make random values
*   Imported *Numpy* library to deal with the values of set

1.   `np.random`: `from numpy import random`  It is the module to make Random Number
2.   `np.random.seed()`: generate random value about the variable of seed in ().
3.  `np.random.uniform()` : function to generate random value according to  uniform distribution in the specific range
4.  `np.random.shuffle()` : function to shutffle the array of the fuction in ().

*   Imported *Matplotlib.pyplot* libaray to visualize data
1.   `figure() `: create new figure
2.   `plot()` : make graph. get value x&y (list)
3.   `show()` : show graph.
4.   `legend` : show the legend (must write 'label' in plot







**1. Generating quadratic function**
- `import time` = time module
- generating variable `NUM_SAMPLES`, initializing as 1000. **The number of how many datas need to be generated?**

- function `np.ramdom.uniform`, importing 1000 numbers of datas in order(`NUM_SAMPLES`) and save into variable `xs`
- function `'np.random.shuffle(xs)'`, shuffle temperary imported xs


**2. Generating quadratic function:more realistical example**

- Add function `np.random.randn(NUM_SAMPLES)`, in order to make more natural curve as normal distribution (Gaussian distribution)
- importing function `np.random.randn(NUM_SAMPLES)`

**3. Separate training and experiment data**
- `NUM_SPLIT = int(0.8*NUM_SAMPLES)` : x,y data -> 800 = training data, and 200 = experiment data
- first 800 of 1000 data is allocated to `x_train`, other 200 data is allocated to  `x_test`, by imported fuction `np.split`

**4. Struct Artificial Neural Network (ANN)**

actual and predicted - before training

- `model_f = tf.keras.Sequentia` function -> From keras library, compose tensor and the number of how many input nodes
- `model_f.compile` function -> compose ANN internally (at least two functions are required. In this case, `loss` function and `optimizer` function is used.)

**5. Train ANN and testing**

actual and predicted - after training

- `model_f.fit` function -> train ANN based on x_train & y_train // how many times of study will be done = epochs
- `model_f.predict` function -> Use Tranined ANN

```python
import time
import numpy as np

import matplotlib.pyplot as plt

NUM_SAMPLES = 1000

np.random.seed(int(time.time()))

xs = np.random.uniform(-2, 0.5, NUM_SAMPLES)
np.random.shuffle(xs)
print(xs[:5])

ys = 2*xs**2 + 3*xs + 3
print(ys[:5])

plt.plot(xs, ys, 'b.')
plt.show()
#--------------- // 1. : Graph - Generating quadratic fuction
ys += 0.1*np.random.randn(NUM_SAMPLES)

plt.plot(xs, ys, 'g.')
plt.show()
#--------------- // 2. : Graph - Generating quadratic fuction using distribution
NUM_SPLIT = int(0.8*NUM_SAMPLES)

x_train, x_test = np.split(xs, [NUM_SPLIT])
y_train, y_test = np.split(ys, [NUM_SPLIT])

plt.plot(x_train, y_train, 'b.', label= 'train')
plt.plot(x_test, y_test, 'r.', label='test')
plt.legend()
plt.show()
#--------------- // 3. : Graph - Separate training and experiment data
import tensorflow as tf

model_f = tf.keras.Sequential(
    [
        tf.keras.layers.InputLayer(input_shape=(1,)),
        tf.keras.layers.Dense(16, activation='relu'),
        tf.keras.layers.Dense(16, activation='relu'),
        tf.keras.layers.Dense(1)
    ]
)

model_f.compile(optimizer='rmsprop',
                 loss='mse')

p_test = model_f.predict(x_test)

plt.plot(x_test, y_test, 'b.', label='actual')
plt.plot(x_test, p_test, 'r.', label='predicted')
plt.legend()
plt.show()
#--------------- // 4. : Graph - Struct Artificial Neural Network (ANN)
model_f.fit(x_train, y_train, epochs=600)

p_test = model_f.predict(x_test)

plt.plot(x_test, y_test, 'b.', label='actual')
plt.plot(x_test, p_test, 'r.', label='predicted')
plt.legend()
plt.show()
#--------------- // 5. : Graph - Train ANN and testing
```

### 4. Deep learning example

> Topic: Train ANN enabling to recognize picture and number



**Setup**

   *   Imported *Tensorflow* library to make ANN, using Keras classes.
1.   `tf.keras.datasets.mnist` module = hand - written data sets. 60 thousands data for learning and 10 thousands data for testing.
2.   `tf.keras.Sequential` : use **Sequential model**  for the most simple ANN (single input & output tensor)
3.   `tf.keras.layers.InputLayer` : this function creates tensor provided by Keras, and set how many inputs are needed (784 Input Layer)
4.   `tf.keras.layers.Dense` : this class make ANN layer. (128 & 10 Hidden Layer)


- `mind.load.data()` function = x_train, x_test -> data, y_train, y_test -> label.
- `x_train / 255.0` -> Because every pics are consisting of 255 pixels, devide every single pixels by 255 in order to transform into the value between 0.0 and 1.0, and put into ANN.
- `model.compile` function -> compose internal ANN. At least two fuction is required.

**4-1 - Make output to express the shape of data**

**4-2 : Show the object figure = enabling pic, imshow = draw the object, show = screen**

**4-3 : draw the value of pixels, 28 x 28**

**4-4 : 5x5 coordinate and draw 10x10 inches pics. camp = color map**

**4-5 : Express from 0 to 1 by divided by 255 (8 bits). Express 60000 data, by consecutive number of 784(28pix^2)**


```python
import tensorflow as tf

mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
print("x_train:%s y_train:%s x_test:%s y_test:%s "%(
      x_train.shape, y_train.shape, x_test.shape, y_test.shape))
#------------------------------------// #4-1 - Make output to express the shape of data
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train, x_test = x_train.reshape((60000, 784)), x_test.reshape((10000, 784))

model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(784,)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=[ 'accuracy' ] )

model.fit(x_train, y_train, epochs=5)

model.evaluate(x_test, y_test)
```
